## Решение систем линейных сравнений (китайская теорема об остатках)

Даны натуральные взаимно простые числа n_1, n_2, ..., n_k и целые остатки r_1, r_2, ..., r_k, а также система

M = r_1 (mod n_1)

M = r_2 (mod n_2)

...

M = r_k (mod n_k)

Мы находим число M, их бесконечно много

В решении используются следующие обозначения:
- `N = n_1*n_2*...*n_k`
- `N_i = N / n_i`
- `inv_i = (N_i)^-1` - обратный элемент для N_i. `inv_i: N_i * M_i = 1 (mod n_i)`

Одно решение: `M = sum(N_i * inv_i * r_i)`

Общее решение: `M = x0 + k * N`, где x0 - минимальное решение; k >= 0, целое

```python
def check_input(n_arr, r_arr):
    """Проверяет корректность введенных данных."""
    if len(n_arr) != len(r_arr):
        raise ValueError("Чисел n и r должно быть одинаковое количество")
    
    for i in range(len(r_arr)):
        if r_arr[i] != int(r_arr[i]):
            raise ValueError(f"r должно быть целым числом, получено {r_arr[i]}")
        if n_arr[i] != int(n_arr[i]):
            raise ValueError(f"n должно быть целым числом, получено {n_arr[i]}")
        if r_arr[i] >= n_arr[i]:
            raise ValueError(f"r (остаток) должно быть меньше n (делитель). Получено r = {r_arr[i]}, n = {n_arr[i]}")
    
    def prime_factorization(n):
        """Возвращает множество простых делителей числа n."""
        res = set()
        while n % 2 == 0:
            n //= 2
            res.add(2)
        div = 3
        while div * div <= n:
            while n % div == 0:
                n //= div
                res.add(div)
            div += 2
        if n > 1:
            res.add(n)
        return res

    seen = set() # множество встреченных делителей
    for n in n_arr:
        for p in prime_factorization(n):
            if p in seen:
                raise ValueError(f"Введенные числа n не являются взаимно простыми. Получено {n_arr}")
            seen.add(p)
```

```python
def solve(n_arr, r_arr):
    """Решает систему уравнений M = r_i (mod n_i), находит решение в общем виде."""
    import math
    check_input(n_arr, r_arr)
    def get_inverse(n, d):
        """Находит обратный элемент n^-1, такой, что n * n^-1 = 1 (mod d)."""
        ans = 0
        while True:
            if (n * ans) % d == 1:
                return ans
            ans += 1
    M = 0
    N = math.prod(n_arr)
    for n, r in zip(n_arr, r_arr):
        N_i = N // n
        inv = get_inverse(N_i, n)
        M += N_i * inv * r
    
    x0 = M % N
    return f"{x0} + {N}k"
```

```python
solve([3, 5, 7], [1, 1, 1])
```

```python
solve([6, 7, 23], [5, 2, 13])
```
