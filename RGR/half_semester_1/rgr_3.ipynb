## Расчетно-графическая работа 3. Таблица Кэли для шестиугольника

```python
import pandas as pd
```

```python
def get_elements(n):
    elements = []
    for i in range(n):
        if i == 0:
            elements.append("e")
        else:
            elements.append(f"a{i}")
    return elements
```

```python
elements = get_elements(6)
cayley_table = pd.DataFrame(index=elements, columns=elements)
cayley_table
```

```python
def fill_table(df: pd.DataFrame):
    for i in range(len(df)):
        for j in range(len(df)):
            element_idx = (i + j) % len(df)
            if element_idx == 0:
                df.iloc[i, j] = "e"
            else:
                df.iloc[i, j] = f"a{element_idx}"
    return df
```

```python
cayley_table = fill_table(cayley_table)
cayley_table
```
