## Вычисление цепных дробей

```python
def calculate(a, b):
    """
    Раскладывает a/b в цепную дробь и возвращает компактный список.
    Общий вид: a/b = x1 + (1 / (x2 + (1 / (x3 + ...))))
    """
    result = []
    while b != 0:
        result.append(a // b)
        a, b = b, a % b
    return result 
```

```python
print(calculate(297, 210))
```

```python
def calc(x):
    """
    Вычисляет периодическую цепную дробь для иррационального числа x.
    Общий вид: x = x1 + (1 / (x2 + (1 / (x3 + ...))))
    Числа [x2, ..., xn] - период дроби
    """
    n = round(x * x)
    a0 = int(x)
    a = a0
    m = 0
    d = 1
    result = [a0]

    while a != 2 * a0:
        m = d * a - m
        d = (n - m * m) / d
        a = int((a0 + m) / d)
        result.append(a)

    return result
```

```python
import math
print(calc(math.sqrt(23)))
```
