## Поиск НОД по алгоритму Евклида и НОК
Вариант 5

```python
# Находим НОД - Greatest Common Divisor
def find_gcd(a, b):
    while b != 0:
        a = a % b    # находим остаток от деления
        a, b = b, a  # меняем местами a и b
    return a

# Находим НОК - Least Common Multiple
def find_lcm(a, b):
    return abs(a * b) / find_gcd(a, b) # |ab| / НОД(a, b)
```

```python
print(f"НОД(24, 8) = {find_gcd(24, 8)}")
print(f"НОК(24, 8) = {find_lcm(24, 8)}")
print()
print(f"НОД(162, 45) = {find_gcd(162, 45)}")
print(f"НОК(162, 45) = {find_lcm(162, 45)}")
print()
print(f"НОД(2341, 72) = {find_gcd(2341, 72)}")
print(f"НОК(2341, 72) = {find_lcm(2341, 72)}")
print()
print(f"НОД(1259, 13) = {find_gcd(1259, 13)}")
print(f"НОК(1259, 13) = {find_lcm(1259, 13)}")
```
