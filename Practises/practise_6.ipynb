## Нахождение n последних цифр для числа x, возведенного в степень pow

Наивный, медленный алгоритм. Сложность: O(p), где p - степень числа. 

```python
def find_n_last_digits(x, pow, n) -> str:
    div = 10 ** n # делитель
    r1 = x % div # остаток от деления x^1 на div
    prev = r1 # предыдущий остаток
    cur = r1 # текущий остаток
    for _ in range(2, pow + 1):
        # текущий остаток равен произведению r1 и предыдущего остатка
        cur = (r1 * prev) % div
        prev = cur

    return str(cur).zfill(n)
        
```

```python
print(find_n_last_digits(2, 10, 3))
print(find_n_last_digits(777, 77, 2))
print(find_n_last_digits(891, 100_000_000, 5))
```

Лучшее, быстрое решение. Сложность: O(log2(p)), где p - степень числа.

Ключевая идея:

Нам нужно найти остаток от деления числа x на 10^n, где n - количество последних цифр

Если мы знаем остаток r от деления числа x^n на d и остаток r1 = x % d, то 
- остаток от деления числа x^(2n) на d будет равен r^2 % d
- остаток от деления числа x^(2n + 1) на d будет равен (r^2 * r1) % d = ((r^2 % d) * r1) % d

Поэтому если раскладывать исходную степень в виде pow = 2 * q + r, то мы можем посчитать нужный нам остаток.

Можно заметить, что для нахождения q и r мы используем тот же метод, что и для нахождения представления числа в двоичном виде.
Пробегаем по двоичной записи, начиная со второй цифры. Если встречаем 0, то это случай с x^(2n), если 1, то x^(2n + 1)

Пример: последние 2 цифры числа 777^77

d = 10^2 = 100; r1 = 777 % 100 = 77 -> 777^1 % 100 = 77.
Представляем степень 77 в двоичном виде: 1001101

1. Начинаем со второй цифры, она равна 0. r = 77^2 % 100 = 29 -> 777^2 % 100 = 29
2. Цифра 0. r = 29^2 % 100 = 41 -> 777^4 % 100 = 41
3. Цифра 1. r = (41^2 * 77) % 100 = 37 -> 777^9 % 100 = 37
4. Цифра 1. r = (37^2 * 77) % 100 = 13 -> 777^19 % 100 = 13
5. Цифра 0. r = 13^2 % 100 = 69 -> 777^38 % 100 = 69
6. Цифра 1. r = (69^2 * 77) % 100 = 97 -> 777^77 % 100 = 97

Ответ: 97.

```python
def get_binary(n):
    """Возвращает представление числа в двоичной системе в виде списка."""
    binary = []
    while n > 0:
        binary.append(n % 2)
        n //= 2
    binary.reverse()
    return binary

def find_n_last_digits(x, pow, n) -> str:
    div = 10 ** n # делитель
    r1 = x % div # остаток от деления x^1 на div
    rem = r1 # динамический остаток, который мы считаем
    binary = get_binary(pow)
    for i in range(1, len(binary)):
        if binary[i] == 0:
            rem = (rem * rem) % div
        else:
            rem = (((rem * rem) % div) * r1) % div
    return str(rem).zfill(n)
```

```python
print(find_n_last_digits(2, 10, 3))
print(find_n_last_digits(777, 77, 2))
print(find_n_last_digits(891, 100_000_000, 5))
print(find_n_last_digits(2, 10**10000, 40)) # 2^(10^10000)
```
