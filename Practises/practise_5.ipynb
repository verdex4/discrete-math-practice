## Решение линейного диафантова уравнения

Алгоритм работает для решения уравнений вида ax + by = c, где
- a, b - произвольные целые коэффициенты
- c = НОД(a, b)

Вспомогательные функции

```python
# Нахождение НОД - Greatest Common Divisor
def find_gcd(a, b):
    while b != 0:
        a = a % b    # находим остаток от деления
        a, b = b, a  # меняем местами a и b
    return a

# Разложение в цепную дробь
def continued_fraction_rational(a, b):
    """
    Раскладывает a/b в цепную дробь и возвращает компактный список.
    Общий вид: a/b = x1 + (1 / (x2 + (1 / (x3 + ...))))
    """
    result = []
    while b != 0:
        result.append(a // b)
        a, b = b, a % b
    return result 
```

```python
def linear_diophantine_equation(a, b, c) -> str:
    """Решает линейное диафантово уравнение с помощью цепных дробей."""
    if c != find_gcd(a, b):
        raise ValueError(f"c = {c} не является НОД({a}, {b}). Решения нет.")
    
    # сохраняем начальные значения
    A = a
    B = b

    frac = continued_fraction_rational(a, b) # цепная дробь
    # отбрасываем последний элемент (по алгоритму)
    a = 1 # числитель
    b = frac[-2] # знаменатель (предпоследний элемент)

    # идем в обратном порядке
    for i in range(len(frac) - 3, -1, -1):
        a = frac[i] * b + a # считаем числитель новой дроби
        a, b = b, a # то же самое, что 1 / (a/b)
    
    x0, y0 = a, -b # находим x0, y0
    # выражаем x, y через x0, y0
    x = f"{x0} + {B}k"
    y = f"{y0} - {A}k"
    return f"x = {x}; y = {y}"
```

```python
print(linear_diophantine_equation(99, 70, 1))
print(linear_diophantine_equation(6, 8, 2))
print(linear_diophantine_equation(17, 5, 1))
```
